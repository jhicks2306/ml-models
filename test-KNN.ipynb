```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from KNN import KNN
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate a dataset.
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, 
                            n_clusters_per_class=1, n_classes=2, random_state=23)
# Split the dataset into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
# Plot the dummay dataset.
feature1_idx = 0
feature2_idx = 1

plt.Figure(figsize=(8,6))

# Scatter plot for each class.
for class_label in range(np.unique(y).size):
    plt.scatter(X[y == class_label, feature1_idx], X[y == class_label, feature2_idx], label=f'Class: {class_label}')

plt.title('Scatter plot of dummy data')
plt.xlabel(f'Feature {feature1_idx + 1}')
plt.ylabel(f'Feature {feature2_idx + 1}')
plt.legend()
plt.show()
```

```python
# Train a KNN classifier and get predictions.
clf = KNN(k=5)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

Finally, let's look at the acuracy of our classifier. 

```python
acc = np.sum(predictions == y_test) / len(y_test)
print(f'Percentage accuracy of KNN classifier: {acc:.2%}')
```
