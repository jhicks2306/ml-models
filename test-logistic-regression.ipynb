```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from LogisticRegression import LogisticRegression
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=2, random_state=23, cluster_std=2.5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=y_test)
```

```python
clf = LogisticRegression(lr=0.001)
```

```python
clf.fit(X_train, y_train)
```

```python
y_preds, _ = clf.predict(X_test)
```

```python
def accuracy(y_preds, y):
    np.sum(y_preds == y) / len(y)
```

```python
acc = accuracy(y_preds, y_test)
print(f'Accuracy of Logistic Regression model: {acc}')
```
